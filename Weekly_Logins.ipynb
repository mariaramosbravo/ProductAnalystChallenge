# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weekly.csv')
df
```

```python
del df["User Name"]

df = df.replace({"No": False, "Yes": True})
df.head()
```

#Box Plots
Some of these boxplots can be useful to estimate which variables have a bigger impact in the number of weekly logins. We can see that the users who login more times are:
*   Users with Slack Integration.
*   Users who's most used page is storyboard.
*   Users that are given free trials.
*   Users from Huron Consulting Group, One Legal, SpaceX, Twilio, Us Soccer Federation (Although the average logins for each company may not be very high, these companies presented maximums in the number of logins per week.)





```python
names = list(df.columns.values)

variables = names[:4] + names[5:]
target = 'Weekly Logins'

fig, ax = plt.subplots(len(variables), figsize=(100,200))
for i, col_val in enumerate(variables):

    sns.boxplot(x=col_val,y=target, data = df, ax=ax[i])

plt.show()
```

# Data Formatting
In order to see the correlation between the variables I formated the data to convert the objects to integer, boolean or float values.

```python
dict_names = {}

for elem in names:
  dict_names[elem] = df[elem].unique()

del dict_names['Slack Integration']
del dict_names['Had Free Trial']
del dict_names['User MRR']
del dict_names['Weekly Logins']
del dict_names['Sales Call in Trial']
del dict_names['Zendesk Integration']
del dict_names['Salesforce Integration']
del dict_names['Gmail Integration']
del dict_names['Sales Touches to Close']
del dict_names['Used Admin Panel']
del dict_names['Has Issued Support Request']
del dict_names['Company Size']
del dict_names['Most common IP']
print(dict_names)

dict_company_size = {
    '>$10B': 20,
    '$500M - $1B': .75,
    '$100M - $500M': .3,
    '<$100M': .05,
    '$1B - $10B': 5
}
df['Company Size'] = df['Company Size'].replace(dict_company_size)
df['Most common IP'].replace("[^a-z0-9_]+", '', regex=True, inplace=True)
df['Most common IP'] = df['Most common IP'].astype(float)


for key, val in dict_names.items():
  dict_temp = {k: v for v, k in enumerate(val)}
  df[key]= df[key].replace(dict_temp)
  #df[key] = pd.Categorical(df[key])

df
```

# Correlation with "Weekly Logins"
I found that the variables that show stronger correlation with the Weekly Logins are "Slack Integration" and "Free Trial". Meaning that the users who have access to slack and are given free trial are the most likely to log in frequently.



```python
df[names[:4] + names[5:]].corrwith(df['Weekly Logins'], method='pearson')
```
